```python
import pandas as pd

data = pd.read_csv("train_extended.csv")

# Calcular correlação entre as variáveis "Age" e "Weight"
correlacao = data[["Age", "Weight"]].corr().iloc[0, 1]
correlacao
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de dispersão (scatter) com as variáveis "Age" e "Weight"
plt.scatter(data["Age"], data["Weight"], alpha=0.5)
plt.xlabel("Idade")
plt.ylabel("Peso")
plt.title("Correlação entre o peso e a idade dos caranguejos")

# Salvar o gráfico como uma imagem
plt.savefig("correlacao_peso_idade.png")
```

```python
# Criar um gráfico de dispersão (scatter) com as variáveis "Length" e "Age"
plt.scatter(data["Length"], data["Age"], alpha=0.5)
plt.xlabel("Tamanho")
plt.ylabel("Idade")
plt.title("Correlação entre o tamanho e a idade dos caranguejos")

# Salvar o gráfico como uma imagem
plt.savefig("correlacao_tamanho_idade.png")
```

```python
# Criar um gráfico de dispersão com as variáveis "Length" e "Age"
plt.hexbin(data["Length"], data["Age"], gridsize=50, cmap="Blues")
plt.xlabel("Tamanho")
plt.ylabel("Idade")
plt.title("Correlação entre o tamanho e a idade dos caranguejos (Hexbin Plot)")

# Adicionar uma barra de cores
cb = plt.colorbar()
cb.set_label("Densidade")

# Salvar o gráfico como uma imagem
plt.savefig("correlacao_tamanho_idade_hexbinplot.png")
```
